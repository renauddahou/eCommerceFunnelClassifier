## Feature Cleaning

```python
# Counting the missing values
print("Null value in customer_state: ", combinedDataset['customer_state'].isnull().sum())
print("Null values in product_category_group: ",combinedDataset['product_category_group'].isnull().sum())
print("Null values in business_type: ",combinedDataset['business_type'].isnull().sum())
print("Null values in origin: ",combinedDataset['origin'].isnull().sum())
print("Null values in lead_behaviour_profile: ",combinedDataset['lead_behaviour_profile'].isnull().sum())
```

Unfortunately, our response variable `lead_behaviour_profile` has the highest number of null values.  
As that's the variable we are trying to predict, its value **cannot be NULL**. As a result, we drop **all null rows**.

```python
# Creating a focused, smaller dataframe
dataset_one = combinedDataset[['customer_state','product_category_group','business_type','origin','lead_behaviour_profile']]

# Dropping all rows with NULL values
dataset_one = dataset_one.dropna(axis=0)

# Converting data types on the columns to categorical
dataset_one['customer_state'] = dataset_one['customer_state'].astype('category')
dataset_one['product_category_group'] = dataset_one['product_category_group'].astype('category')
dataset_one['business_type'] = dataset_one['business_type'].astype('category')
dataset_one['origin'] = dataset_one['origin'].astype('category')
dataset_one['lead_behaviour_profile'] = dataset_one['lead_behaviour_profile'].astype('category')

print('Dataset Information: \n')
print(dataset_one.info())
dataset_one.dtypes
```

```python
# Label encoding to generate numerical values for the neural network
dataset_one['customer_state_numeric'] = pd.factorize(dataset_one['customer_state'], sort=True)[0]
dataset_one['product_category_group_numeric'] = pd.factorize(dataset_one['product_category_group'], sort=True)[0]
dataset_one['business_type_numeric'] = pd.factorize(dataset_one['business_type'], sort=True)[0]
dataset_one['origin_numeric'] = pd.factorize(dataset_one['origin'], sort=True)[0]
dataset_one['lead_behaviour_profile_numeric'] = pd.factorize(dataset_one['lead_behaviour_profile'], sort=True)[0]

dataset_one.head()
```

# Machine Learning Models

### Split the data into train/test sets

```python
# Train and test sets with numeric values
from sklearn.model_selection import train_test_split

X = pd.DataFrame(dataset_one[['customer_state_numeric','product_category_group_numeric','business_type_numeric','origin_numeric']])
y = dataset_one[['lead_behaviour_profile_numeric']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=7, test_size=0.20)

# creating another copy for multivariate linear regression
X_train_lin, X_test_lin, y_train_lin, y_test_lin = train_test_split(X, y, random_state=7, test_size=0.20)

print('X:')
print(X.info())
print('\ny:')
print(y.info())
```

```python
# Train and test sets with categorical values
from sklearn.model_selection import train_test_split

X_cat = pd.DataFrame(dataset_one[['customer_state','product_category_group','business_type','origin']])
y_cat = dataset_one[['lead_behaviour_profile']]

X_train_cat, X_test_cat, y_train_cat, y_test_cat = train_test_split(X_cat, y_cat, random_state=7, test_size=0.20)


print('X:')
print(X_cat.info())
print('\ny_cat:')
print(y_cat.info())
```

## Implement our models  

a) Neural Net   
b) K-Nearest Neighbour  
c) Multivariate Linear Regression 

We implemented these three machine learning models to compare their performance and see which is the best.

#### K-Nearest Neighbours 

```python
# import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier

#Create KNN Classifier
k=4
knn = KNeighborsClassifier(n_neighbors=k)

#Train the model using the training sets
knn.fit(X_train, y_train.values.ravel())

#Predict the response for test dataset
y_hat = knn.predict(X_test)

#accuracy
from sklearn import metrics
print('Train set accuracy: ',metrics.accuracy_score(y_train, knn.predict(X_train)))
print('Test set accuracy: ',metrics.accuracy_score(y_test, y_hat))
```

### Neural Net

```python
# Neural Network Model
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Dense(4),
                                    tf.keras.layers.Dense(128, activation='relu'),
                                    tf.keras.layers.Dense(5, activation='softmax')
])

model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy','categorical_accuracy'])

# Train the neural network
history = model.fit(X_train, y_train, epochs=50)


# Plot metrics
plt.plot(history.history['accuracy'],
         color='#FEE500'
         )
plt.show()
print()
```

### Multi-variate regression

```python

# import essential models and functions from sklearn
from sklearn.metrics import mean_squared_error
from sklearn import linear_model


linreg = linear_model.LinearRegression()


linreg.fit(X_train_lin, y_train_lin)        

# Coefficients of the Linear Regression line
print('Intercept of Regression \t: b = ', linreg.intercept_)
print('Coefficients of Regression \t: a = ', linreg.coef_)
print()


# Check the Goodness of Fit (on Train Data)
y_train_lin_hat = linreg.predict(X_train_lin)
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t: %.5f " % linreg.score(X_train_lin, y_train_lin))
print("Mean Squared Error (MSE) \t: %.5f " % mean_squared_error(y_train_lin, y_train_lin_hat))
print()

# Check the Goodness of Fit (on Test Data)
y_test_lin_hat = linreg.predict(X_test_lin)
print("Goodness of Fit of Model \tTest Dataset")
print("Explained Variance (R^2) \t: %.5f " % linreg.score(X_test_lin, y_test_lin))
print("Mean Squared Error (MSE) \t: %.5f " % mean_squared_error(y_test_lin, y_test_lin_hat))
print()
#plt.plot(y_train_lin, y_train_lin_hat, 'o')
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

clf = RandomForestClassifier(n_estimators = 100)
clf.fit(X_train_lin, y_train_lin.values.ravel())

print("Evaluation Metrics \t")
y_hat = clf.predict(X_test_lin)
print()
print("Accuracy score : %.3f " % accuracy_score(y_test_lin, y_hat))
print("Confusion Matrix : \n" , confusion_matrix(y_test_lin, y_hat))
```
